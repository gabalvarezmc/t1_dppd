## 3. Visualización y análisis

Se puede observar que el puntaje f1 obtenido comienza muy bajo en febrero y marzo, aumentando abruptamente para abril y bajando constantemente hasta noviembre. Esto puede deberse a diferencias en el comportamiento de las personas en estos meses.



Por otro lado, se puede observar que la cantidad de personas que dieron propinas altas bajó abruptamente luego de marzo. Un posible motivo es el efecto de la pandemia en el comportamiento de los pasajeros.



Además, se puede observar que los valores predichos tienen un comportamiento similar a los datos reales.



Es importante considerar le disminución de viajes por temas de encierro por pandemia luego del abril.


## 4. Análisis crítico y explicación de resultados

El rendimiento del modelo no es consistende en los diferentes meses evaluados. Esto puede deberse a diferentes factores: 
- Cambios de comportamiento de los pasajeros por motivos de pandemia o por motivos económicos por la crisis originada por esta misma.
- Existencia de estacionalidad en el comportamiento anual de los pasajeros. Para comprobar este punto se debería comparar datos de diferentes años para encontrar patrones.

Para mejorar la robustez del modelo, se debería entrenar con una mayor cantidad de datos y con años con comportamientos similares a los que se esperan evaluar. Debe también considerarse la inclusión de otros campos predictores que puedan apoyar los resultados.
